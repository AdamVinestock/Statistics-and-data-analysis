```python
# Example code for bivariate pdfs
from scipy.stats import multivariate_normal as mn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def plot_2d_Gaussian_pdf(means, cov, header):
    n = 100
    x1 = np.linspace(means[0] - 3 * np.sqrt(cov[0][0]), means[0] + 3 * np.sqrt(cov[0][0]), n)
    x2 = np.linspace(means[1] - 3 * np.sqrt(cov[1][1]), means[1] + 3 * np.sqrt(cov[1][1]), n)
    x1_v, x2_v = np.meshgrid(x1, x2)
    Xgrid = np.vstack([x1_v.ravel(), x2_v.ravel()]).T
    Y = mn.pdf(Xgrid, means, cov)
    fig, ax = plt.subplots()
    ax.pcolorfast(x1, x2, Y.reshape(x1_v.shape), alpha=0.5, cmap='Blues')
    ax.contour(x1_v, x2_v, Y.reshape(x1_v.shape), alpha=0.3, colors='b')
    ax.grid(alpha=0.2)
    plt.suptitle(header)
    plt.show()
    
means = [3, 2]
cov = [[1, 0.5], [0.5, 0.8]]
plot_2d_Gaussian_pdf(means, cov,"example")
```

* Using confidence intervals determine for which numerical features you can state with confidence of 95% that the healthy population (target = 0) mean is larger/smaller than the disease population (target = 1) mean. 

```python
from scipy import stats
import math

print("We create a delta of E(healthy_attribute)-E(diseased__attribute) and examine if CI is positive or negative in it's whole: " + "\n")
df_healthy = df.loc[df['target'] == 0]
df_diseased = df.loc[df['target'] == 1]
gamma = stats.norm.ppf(1-0.05/2)
def compare_conf_interval (df_h, df_d,attribute, gamma): #returns nested tuple of conf intervals
    h_mean = df_h[attribute].mean()
    d_mean = df_d[attribute].mean()
    delta = h_mean-d_mean
    margin = gamma*math.sqrt(df_h[attribute].std()**2/(len(df_h))+df_d[attribute].std()**2/(len(df_d)))
    interval = (round(delta-margin,2),round(delta+margin,2))
    return interval

age = compare_conf_interval (df_healthy,df_diseased,'age',gamma)
print("Delta mean age interval: {}".format(age))
print("Using 95% CI, mean age in healthy population is higher.".format(age[1]) +"\n")

trestbps = compare_conf_interval (df_healthy,df_diseased,'trestbps',gamma)
print("Delta mean resting blood pressure interval: {}".format(trestbps))
print("Using 95% CI, mean blood pressure in healthy population is higher."+"\n")

chol = compare_conf_interval (df_healthy,df_diseased,'chol',gamma)
print("Delta mean cholestoral interval: {}".format(chol))
print("Using 95% CI we cannot determine higher/lower cholestoral in healthy population."+"\n")

thalach = compare_conf_interval (df_healthy,df_diseased,'thalach',gamma)
print("Delta max hr achieved interval: {}".format(thalach))
print("Using 95% CI, max hr achieved in healthy population is lower."+"\n")

oldpeak = compare_conf_interval (df_healthy,df_diseased,'oldpeak',gamma)
print("Delta mean depression induced by exercise interval: {}".format(oldpeak))
print("Using 95% CI, mean depression induced by exeercise in healthy population is higher."+"\n")
```

### sECtion 3 - Heavy Tailed Distributions and the Exponential Distribution

### Heavy Tailed Distributions

Recall the definition of Heavy Tailed distribution from the lectures.

*A distribution is said to have a heavy right tail if its tail probabilities vanish slower than any exponential*
$$ \forall t>0, \lim_{x\to\infty} e^{tx}P(X>x)=\infty $$

Does the standard log-normal distribution have a heavy right tail? prove your answer.

Let X be the standart log normal rv.
We showed in class that : 
$$P(X<=x) = \Phi(\ln{x})$$

Therefore the limit is : 
$$\lim_{x\to\infty} e^{tx}P(X > x) = \lim_{x\to\infty} e^{tx}*(1- \Phi_{ln(x)}) = \lim_{x\to\infty} \frac{(1- \Phi_{ln(x)})}{e^{-tx}} =  \$$

Apply L'Hôpital rule and caclulate the derivatives of both functions 
$$ \lim_{x\to\infty} \frac{(- \varphi {ln(x)})}{-te^{-tx}}  = $$
$$ \lim_{x\to\infty} \frac{1}{t * \sqrt{2 \pi}} * e^{tx - \frac{1}{2}(ln{x})^{2}} = \infty $$


Therefor, the standart log normal has a heavy right tail.

### Special Properties of the Exponential Distribution

Let $X_1 \sim exp(\lambda_1)$ and $X_2 \sim exp(\lambda_2)$ be two independent exponential random variables.

Calculate $P(X_1 < X_2)$.

$$ P(X_1 < X_2) = \int_{0}^{\infty} P(X_1 = x)P(X_2 > x) dx  	= $$
$$ = \int_{0}^{\infty} f_1(x) * (1- F_{2}(x)) dx  	= $$
 $$ = \int_{0}^{\infty} \lambda_1 e ^{-\lambda_1x} * (1 - ( 1 - e^{- \lambda_2x}))  dx  	= $$
 $$ = \int_{0}^{\infty} \lambda_1 e ^{-\lambda_1x} * e^{-\lambda_2x} dx = $$
 $$ = \lambda_1 * \int_{0}^{\infty} e ^{-x(\lambda_1 + \lambda_2)} dx = $$
 $$ = \lambda_1 * (\frac{1}{\lambda_1 + \lambda_2})= $$
 $$ = \frac{\lambda_1}{\lambda_1 + \lambda_2 } $$

### Section 4 - CLT for Markov chains

In this exercise you will construct trajectories of Markovian dice rolling results in the following way.<br>
The first roll, X0, is Unif(1..6)<br>
After i rolls are determined the i+1st, Xi+1, is drawn according to the row that corresponds to the value of Xi in the matrix T below. <br>
In other words, T is the transition matrix of a Markov chain and the initial distribution is uniform.

\begin{equation*}

1. Notice that $ \sigma = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] $ is a stationary distribution over states {1,2,3,4,5,6} since : 
$$ \sigma * T = \sigma $$
$$ \sigma = \pi_{0} = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] $$
because: $$ \pi_{1} = \pi_{0} * T = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] =  \sigma * T  $$
$$	\downarrow$$
$$ \forall n,  \pi_{n} = \sigma $$

Therefore, the probability vector is identical in all steps and each is the unified distribution over the 6 states.
For this reason, I will expect the average of all trajectories to behave like the unified distribution and be $$\mu =  \frac{6+1}{2} = 3.5 $$


```python
import numpy as np

T = np.array([
[ 0.4, 0.2, 0.1, 0, 0.1, 0.2 ],
[ 0.2, 0.4, 0.2, 0.1, 0, 0.1 ],
[ 0.1, 0.2, 0.4, 0.2, 0.1, 0 ],
[ 0, 0.1, 0.2, 0.4, 0.2, 0.1 ],
[ 0.1, 0, 0.1, 0.2, 0.4, 0.2 ],
[ 0.2, 0.1, 0, 0.1, 0.2, 0.4 ]]);

# Calculates the vector of probabilities of step i+1 given the ith state
def getNextProbabilitiesVector(T, prevState):
        
    # create Xi vector with 1 at the chosen state index (i1 since the index of state 1 is 0), and 0s elsewhere
    # example : if the sampled value was 1, xi = [1,0,0,0,0,0]
    xi = np.zeros(6)
    xi[prevState - 1] = 1;
        
    return xi.dot(T);

def GenerateTrajectory(size = 30, T = T, pi_0=np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])):
    pi_i = pi_0
    trajectory = np.zeros(size)

    for i in range(size):

        # chose next state from current distribution
        si = np.random.choice([1,2,3,4,5,6], 1, p=pi_i);
        trajectory[i] = si;

        # calculate P(Xi+1 | Xi = si)
        pi = getNextProbabilitiesVector(T, si)
    
    return trajectory;
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# create all averages of 1000 trajectories
empiricAverages = [np.average(GenerateTrajectory()) for i in range(1000)]

plt.figure(figsize=(5, 4))
plt.hist(empiricAverages, density=True, alpha=0.6, bins = 20)
plt.title('Averages of 1000 trajectories of length 30 each')
plt.xlabel('Average')

# Plot the PDF.
mu, std = norm.fit(empiricAverages) 

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std) 
plt.plot(x, p, linewidth=2)

plt.show()
```

The histogram looks like a normal distribution.

#### 4.D
Write a python code to calculate $Cov(X_0,X_i)$ for $i=1,..,500$ based on $\pi_0$ and $T$. 

Compute the values for our case.

```python
# import numpy as np;
from numpy.linalg import matrix_power
from scipy import stats

def GetPi_i_Given_x0(x0, T, i):
    pi_0 = np.zeros(6);
    pi_0[x0] = 1;
    return pi_0.dot(matrix_power(T, i));

# returns matrix M which represents the joint probability function of (X0, Xi), such that : M[j,k] = P(X0 = j + 1 && Xi = k + 1)
def CreateJointDistribution(pi_0, T, i):
    jointDist = np.zeros(shape=(6,6));

   # for each possible value j of X0, calculae the the probability of each possible value k for X1
    for j in range(0,6):
        
        # calculate the vector probabilities of xi given that x0 = j
        pi_i = GetPi_i_Given_x0(x0=j, T=T, i=i);
        for k in range (0, 6):
            jointDist[j,k] = pi_0[j] * pi_i[k];
            
    return jointDist;

# calculates the covariance of x0, and xi given T and Pi_0
def CalculateConvariance(pi_0, T, i):
    values = np.arange(6) + 1
    x0_dist = stats.rv_discrete(name='x0', values=(values, pi_0))
    
    # calculate x_i distribution
    pi_i = pi_0.dot(matrix_power(T, i))
    xi_dist = stats.rv_discrete(name='xi', values=(values, pi_i))
    
    mu_0 = x0_dist.mean();
    mu_i = xi_dist.mean()
    
    # calculate the joint probability distribution
    jointDist = CreateJointDistribution(pi_0, T, i)
    
    # calculate the covariance (based on the formula: cov(x,y) = E[(X- E(X))(Y - E(Y))]
    cov = 0;
    for j in range(1,7):
        for k in range (1, 7):
            # retreive P(x0 = j && xi = k) from the joint distribution matrix (we deduct -1 from both indexes since P(x0=1 && xi=1) is stored in jointDist[1-1][1-1])
            pj_k = jointDist[j-1][k-1]
            cov += pj_k * (j - mu_0)*(k - mu_i)
    return cov;


def CalculateAllConvariancesUpToXi(pi_0, T, i):
    allCovs = [CalculateConvariance(pi_0, T, j) for j in range (1,i)];
    return allCovs;
```

```python
### In our case:

pi_0=np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6]);
T = np.array([
[ 0.4, 0.2, 0.1, 0, 0.1, 0.2 ],
[ 0.2, 0.4, 0.2, 0.1, 0, 0.1 ],
[ 0.1, 0.2, 0.4, 0.2, 0.1, 0 ],
[ 0, 0.1, 0.2, 0.4, 0.2, 0.1 ],
[ 0.1, 0, 0.1, 0.2, 0.4, 0.2 ],
[ 0.2, 0.1, 0, 0.1, 0.2, 0.4 ]]);

CalculateAllConvariancesUpToXi(pi_0, T,500)
```
